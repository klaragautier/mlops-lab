```python
import pandas as pd
```

```python
# Read a comma-separated values (csv) file into DataFrame using `read_csv` method
filepath = "../datastores/raw_csv_data/census.csv"

df = pd.read_csv(filepath)
```

```python
## Return the first n rows using `head` method
# your code here
df_head = df.head(5)
print(df_head)
```

```python
# Returns the last n rows using `tail` method
# your code here
df_tail = df.tail(5)
print(df_tail)
```

```python
# Print a concise summary of a DataFrame using ìnfo` method
## your code here
df.info()
```

```python
# Retrun the column labels of the DataFrame using `columns` method
## your code here
df_columns = df.columns
print(df_columns)
```

```python
#  removes any leading or trailing whitespace from all column names in the DataFrame 
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
# check the updated column names
# your code here
print(df.columns)
```

```python
# Return number of duplicate rows using `duplicated` and `sum` method
## your code here
num_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
# Return DataFrame with duplicate rows removed using `drop_duplicates` method
## your code here
df_no_duplicates = df.drop_duplicates()
print(f"DataFrame shape after removing duplicates: {df_no_duplicates.shape}")
```

```python
# Detect missing values : return number missing for each column in dataframe
# by using ìsna` and `sum` method
## your code here
missing_values = df.isna().sum()
print("Missing values in each column:")
print(missing_values)
```
